```python
%reset -f
import os
import requests

# フォルダが存在しなければ作成
dir = 'data/download/'
if not os.path.isdir(dir):
  os.makedirs(dir)

fn = 'akaoni.png'
path = 'akaoni_dot.png'
res = requests.get(path)

if res.status_code != 200:
  raise Exception(f'ファイル {path} の取得に失敗。強制終了します。Code:{res.status_code}')
else :
  with open(f'{dir}/{fn}','wb') as file: 
    file.write(res.content)
  print(f'{dir} に {fn} を保存しました')
```
